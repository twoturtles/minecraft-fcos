```python
import logging
from pathlib import Path
import random

import ultralytics as ul

import bb
import tt

LOG = logging.getLogger(__name__)
tt.logging_init()
```

```python
data_path = Path.home() / "src/data"
mc_data_path = data_path / "minecraft/mobs/info.json"
yolo_path = data_path / "yolo"
models_path = yolo_path / "models"
best = Path("/Users/joe/src/data/yolo/runs/detect/keep1/weights/best.pt")

ul.settings.update(
    {
        "datasets_dir": str(yolo_path / "datasets"),
        "weights_dir": str(yolo_path / "weights"),
        "runs_dir": str(yolo_path / "runs"),
    }
)
```

```python
dset = bb.Dataset.load(mc_data_path)
model = ul.YOLO(best)
```

```python
# Display results of one query at a time
def detect_single(ir: bb.ImageResult):
    pred = model(ir.full_path)[0]
    return bb.yr_to_ir(pred)


bb.InferViewer[bb.ImageResult](
    detect_single, dset.images, dset.categories
).show_widget()
```
