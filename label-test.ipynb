```python
# Assuming you have file_list defined
# file_list = ['file1.jpg', 'file2.jpg', 'file3.jpg', ...]

color_test = [
            "#1f77b4",
            "#ff7f0e",
            "#2ca02c",
            "#d62728",
            "#9467bd",
            "#8c564b",
            "#e377c2",
            "#7f7f7f",
            "#bcbd22",
            "#17becf",
            "#1f77b4",
            "#ff7f0e",
            "#2ca02c",
            "#d62728",
            "#9467bd",
            "#8c564b",
            "#e377c2",
            "#7f7f7f",
            "#bcbd22",
            "#17becf",
]


file_list = files.copy()

# Create the slider
slider = widgets.IntSlider(
    value=0,
    min=0,
    max=len(file_list) - 1,
    step=1,
    description='Index:',
    continuous_update=False
)

# Create buttons
back_button = widgets.Button(
    description='Back',
    button_style='warning',
    icon='arrow-left'
)

submit_button = widgets.Button(
    description='Submit',
    button_style='success',
    icon='check'
)

skip_button = widgets.Button(
    description='Skip',
    button_style='warning',
    icon='arrow-right'
)

w_bbox = BBoxWidget(
    image = str(file_list[0]),
    classes=CATEGORIES,
    colors=color_test,
    hide_buttons=True
)

out = widgets.Output(
    layout={
        "height": "200px",  # Fixed height
        "overflow": "auto",  # Enables scrollbar
        "border": "1px solid black",
    }
)


def set_bbox(index: int) -> None:
    w_bbox.image = str(files[index])
    w_bbox.bboxes = [] 

# Define callback for slider changes
def on_slider_change(change):
    # This is called whenever the slider value changes
    new_index = change['new']
    with out:
        print(f"Slider moved to index {new_index}: {file_list[new_index]}")
        print(change)
    # Add any other logic you need when the slider moves
    # (e.g., update display, load new image, etc.)
    set_bbox(new_index)

# Define your callback functions
def on_submit(button):
    # Your submit logic here
    display(f"Submitted index {slider.value}: {file_list[slider.value]}")
    
    # Move to next item
    if slider.value < len(file_list) - 1:
        slider.value += 1
    else:
        display("Reached end of list")

def bbox_skip():
    on_skip(skip_button)
def bbox_submit():
    on_submit(submit_button)

w_bbox.on_skip(bbox_skip)
w_bbox.on_submit(bbox_submit)


def on_back(button):
    # Your skip logic here
    display(f"Back index {slider.value}: {file_list[slider.value]}")
    
    # Move to next item
    if slider.value > 0:
        slider.value -= 1
    else:
        display("Reached beginning of list")

def on_skip(button):
    display(f"BUTTON {button}")
    # Your skip logic here
    display(f"Skipped index {slider.value}: {file_list[slider.value]}")
    
    # Move to next item
    if slider.value < len(file_list) - 1:
        slider.value += 1
    else:
        display("Reached end of list")

# Connect slider to observer
slider.observe(on_slider_change, names='value')

# Connect buttons to callbacks
back_button.on_click(on_back)
submit_button.on_click(on_submit)
skip_button.on_click(on_skip)

# Layout the widgets
button_box = widgets.HBox([back_button, submit_button, skip_button])
ui = widgets.VBox([slider, button_box, w_bbox, widgets.Label("Debug Output:"), out])

print(f"classes {w_bbox.classes}")

# Display
display(ui)
```
