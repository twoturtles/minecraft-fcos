```python
import logging
import json
from pathlib import Path

from jupyter_bbox_widget import BBoxWidget
import ipywidgets as widgets


import tt

LOG = logging.getLogger(__name__)
tt.logging_init()
```

```python
CATEGORIES = [
    "chicken",
    "cow",
    "creeper",
    "enderman",
    "pig",
    "sheep",
    "skeleton",
    "spider",
    "villager",
    "zombie",
]
CATEGORIES.sort()

mobs_path = Path.home() / "src/data/mobs1"
images_path = mobs_path / "640"
```

```python
files = sorted(list(images_path.glob("*.png")))
annotations = {}
annotations_path = mobs_path / 'bbw.json'

# a progress bar to show how far we got
w_progress = widgets.IntProgress(value=0, max=len(files), description='Progress')
# the bbox widget
w_bbox = BBoxWidget(
    image = str(files[0]),
    classes=CATEGORIES
)

# combine widgets into a container
w_container = widgets.VBox([
    w_progress,
    w_bbox,
])
```

```python
def skip():
    w_progress.value += 1
    # open new image in the widget
    w_bbox.image = str(files[w_progress.value])
    # here we assign an empty list to bboxes but 
    # we could also run a detection model on the file
    # and use its output for creating inital bboxes
    w_bbox.bboxes = [] 

# when Submit button is pressed we save current annotations
# and then move on to the next file
def submit():
    image_file = str(files[w_progress.value])
    # save annotations for current image
    annotations[image_file] = w_bbox.bboxes
    with open(annotations_path, 'w') as f:
        json.dump(annotations, f, indent=4)
    # move on to the next file
    skip()

w_bbox.on_skip(skip)
w_bbox.on_submit(submit)
```
